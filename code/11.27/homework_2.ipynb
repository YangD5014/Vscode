```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
def year_to_seconds(years:int|float)->float|int:
    seconds = 365*24*60*60*years
    return seconds
```

```python
def P_decayheat_1():
    P_th = 1380
    T = year_to_seconds(years=3)
    T_lasting = year_to_seconds(years=1)
    result = []
    time = []
    for i in range(1000,T_lasting,1000):
        tmp = P_th*0.066* ((i)**(-0.2) - (T+i)**(-0.2))
        result.append(tmp)
        time.append(i)
    return (result,time)

def P_decayheat_2():
    P_th = 2381
    T = year_to_seconds(years=3)
    T_lasting = year_to_seconds(years=1)
    result = []
    time = []
    for i in range(1000,T_lasting,1000):
        tmp = P_th*0.066* ((i)**(-0.2) - (T+i)**(-0.2))
        result.append(tmp)
        time.append(i)
    return (result,time)

def P_decayheat_3():
    P_th = 2381
    T = year_to_seconds(years=0.5)
    T_lasting = year_to_seconds(years=0.5)
    T_lasting = int(T_lasting)
    result = []
    time = []
    for i in range(1000,T_lasting,1000):
        tmp = P_th*0.066* ((i)**(-0.2) - (T+i)**(-0.2))
        result.append(tmp)
        time.append(i)
    return (result,time)
```

### Simulation for One year of decay heat

```python
result1 = P_decayheat_1()
result2 = P_decayheat_2()
result3 = P_decayheat_3()
fig = plt.figure(figsize=(18,5))
plt.subplot(1, 3, 1)
plt.plot(result1[1],result1[0])
plt.title("Reactor 1")
plt.xlabel('Lasting time [seconds]')
plt.ylabel('Decy heat[MW]')
plt.subplot(1, 3, 2)
plt.plot(result2[1],result2[0])
plt.title("Reactor 2")
plt.xlabel('Lasting time [seconds]')
plt.ylabel('Decy heat[MW]')
plt.subplot(1, 3, 3)
plt.plot(result3[1],result3[0])
plt.title("Reactor 3")
plt.xlabel('Lasting time [seconds]')
plt.ylabel('Decy heat[MW]')
plt.show()
```
